Определите доверительный интервал для математического ожидания на уровне $\alpha$ (для каждого $\alpha$ из списка) c использованием...

```python
alpha_list = np.array([0.1, 0.075, 0.05, 0.025, 0.01]) 
```

Распределения Стьюдента

```python
def student_mci(data, alpha=0.05):
    n = data.shape[0]
    mu = data.mean()
    sigma = (((data - mu)**2).sum()/(n-1))**0.5
    t_distr = sp.stats.t(n-1)
    t = t_distr.ppf(alpha/2)
    add = sigma/(n**0.5)*t
    return mu+add, mu-add
```

Бутстрэп оценивания

```python
def bootstrap_mci(data, alpha=0.05):
    B = 10000
    res = np.random.choice(X, size=(B, X.shape[0]))
    res = np.sort(res.mean(axis=-1))
    a = res[int(B*alpha/2)]
    b = res[int(B*(1-alpha/2))]
    return a,b
```

Сравните результаты и сделайте выводы.

```python
res = defaultdict(list)
res["alpha"] = alpha_list
res["Student"] = [np.round(student_mci(X, alpha), 3) for alpha in alpha_list]
res["Bootstrap"] = [np.round(bootstrap_mci(X, alpha), 3) for alpha in alpha_list]
res = pd.DataFrame(res)
res
```

```python
res = defaultdict(list)
res["alpha"] = alpha_list
res["Student"] = [np.round(student_mci(X, alpha), 3)[1]\
                  - np.round(student_mci(X, alpha), 3)[0] for alpha in alpha_list]
res["Bootstrap"] = [np.round(bootstrap_mci(X, alpha), 3)[1] -\
                     np.round(bootstrap_mci(X, alpha), 3)[0] for alpha in alpha_list]
res = pd.DataFrame(res)
res
```

**Выводы:** Оба метода показали похожие результаты, хоть и работают совершенно по-разному. Доверительные интервалы, полученные методом Стьюдента шире.

Оцените доверительный интервал для показателя *ACCURACY* с использованием...

Метода Клоппера-Пирсона (с использованием аппроксимации через бета-распределение)

```python
def bin_clopper_pearson(data, alpha=0.05):
    x = data.sum()
    n = data.shape[0]
    q1 = sp.stats.f(2*x,2*(n-x+1)).ppf(alpha/2)
    q2 = sp.stats.f(2*(x + 1),2*(n-x)).ppf(1 - alpha/2)
    res1 = (1 + (n - x + 1)/(x*q1))**(-1)
    res2 = (1 + (n - x)/((x+1)*q2))**(-1)
    return res1, res2
```

Нормальной аппроксимации

```python
def bin_normal_approx(data, alpha=0.05):
    s = data.sum()
    n = data.shape[0]
    f = n-s
    z = sp.stats.norm().ppf(1-alpha*0.5)
    add = z/n*(s*f/n)**0.5
    return s/n - add, s/n + add
```

Сравните результаты и сделайте выводы.

```python
res = defaultdict(list)
res["alpha"] = alpha_list
res["Clopper-Pearson"] = [np.round(bin_clopper_pearson(Y, alpha), 3) for alpha in alpha_list]
res["Normal"] = [np.round(bin_normal_approx(Y, alpha), 3) for alpha in alpha_list]
res = pd.DataFrame(res)
res
```

```python
res = defaultdict(list)
res["alpha"] = alpha_list
res["Clopper-Pearson"] = [np.round(bin_clopper_pearson(Y, alpha), 3)[1] - \
                          np.round(bin_clopper_pearson(Y, alpha), 3)[0] for alpha in alpha_list]
res["Normal"] = [np.round(bin_normal_approx(Y, alpha), 3)[1] - \
                 np.round(bin_normal_approx(Y, alpha), 3)[0] for alpha in alpha_list]
res = pd.DataFrame(res)
res
```

**Выводы:** Метод Клоппера-Пирсона дает слегка более широкие доверительные интервалы.

### Задача 3

Два предприятия выпускают  изделия одного и того же назначения. Предприятие I выпускает изделия с математическим ожиданием срока службы $T_1$, предприятие II выпускает изделия с математическим ожиданием срока службы $T_2$. Сроки службы изделий обоих предприятий распределены нормально и стандартные отклонения для них равны $\sigma_1$ и $\sigma_2$ соответственно. При этом $T_1 < T_2$, но изделия второго предприятия дороже.

```python
T1, T2, sigma1, sigma2 = 10, 16, 2, 3

def plot_T(T, sigma, c, label):
    norm = sp.stats.norm(loc=T, scale=sigma)
    x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 1000)
    plt.plot(x, norm.pdf(x), c=c)
    plt.vlines(T, *plt.ylim(), linestyles='--', colors=c, label=label)
    
plot_T(T1, sigma1, 'b', 'T1')
plot_T(T2, sigma2, 'r', 'T2')
plt.legend()
plt.show()
```

Третье предприятие утверждает, что ему удалось разработать технологию производства изделий со сроком службы  предприятия II, но по цене предприятия I. В качестве доказательства предлагаются результаты тестирования для $k$ изделий. 

Нулевая гипотеза $H_0$ предполагает, что на самом деле срок службы изделий предприятия III имеет распределение, совпадающее с распределением срока службы изделий предприятия I. Альтернативная гипотеза $H_1$ предполагает, что срок службы изделий предприятия III имеет распределение, совпадающее с распределением срока службы изделий предприятия II. Проверка этих двух предположений производится путем сравнения эмпирически оцененного среднего срока службы протестированных изделий предприятия III с некоторым порогом $T$. Конкретное значение $T$ находится из условия того, что допустимая ошибка первого рода равна $\alpha$.

```python
def sample_from_T3(k):
    "Draws k samples from the 3rd company products lifetime distribution."
    T3, sigma3 = 14, 2.5  # Ooops, spoiler...
    return T3 + sigma3 * np.random.randn(k)  

alpha = 0.05
```

Постройте график мощности теста в зависимости от $k$ и сделайте выводы.

```python
threshold = lambda k, alpha: sp.stats.norm(loc=T1, scale=sigma1/(k**0.5)).ppf(1-alpha)
```

```python
power = lambda k, alpha: 1-sp.stats.norm(loc=T2, scale=sigma2/(k**0.5)).cdf(threshold(k, alpha))
```

```python
power_list = [power(i+1, alpha) for i in range(10)]
plt.plot(np.arange(10)+1, power_list)
plt.xlabel('k')
plt.ylabel('power')
```

**Выводы:** Мощность - вероятность принятия альтернативной гипотезы при условии что альтернативная гипотеза верна. Решение о принятии альтернативной гипотезы принимается основываясь на сравнении эмпирической оценки интересующего нас значения $Y$ с порогом $T$. Порог выбирается так, чтобы свести к допустимому минимуму вероятность ошибки первого рода, т.е отклонить верную нулевую гипотезу. В нашем случае вероятность ошибки первого рода $ 5\% $ и при этом значение мощности очень близко к $1$, значит критерий достаточно хорошо должен справиться с такой задачей.
